```python
def histogram_plotter(title, data1, data2):
    sns.set(rc={"figure.figsize": (10, 7)})
    sns.set_style('white')
# data1 input
    dist = sns.distplot(data1, hist_kws={'alpha':0.2}, kde_kws={'linewidth':5})
    
    dist.axvline(np.mean(data1), lw=5, c='darkred', ls='dashed')
    average_roll = '{0:.2f}'.format(np.mean(data1))
    title = 'Distribution of ' + title + '; average = '  + average_roll + '\n'
    dist.set_title(title, fontsize=16)
    dist = sns.distplot(data1, hist_kws={'alpha':0.2}, kde_kws={'linewidth':5})
    
# data2 input
    dist = sns.distplot(data2, hist_kws={'alpha':0.2}, kde_kws={'linewidth':5})
    
    dist.axvline(np.mean(data2), lw=5, c='darkblue', ls='dashed')
    average_roll = '{0:.2f}'.format(np.mean(data2))
    title = '' + title + '; average = '  + average_roll + '\n'
    dist.set_title(title, fontsize=16)
    dist = sns.distplot(data2, hist_kws={'alpha':0.2}, kde_kws={'linewidth':5})
    
# Creating function to do a histogram plotter for two variables 
```

```python
# Calculate the mean and standard deviation for number of achieved goals by treatment group and gender

mean_total_co_day = df[df['TreatmentGroup']!=1].groupby(['TreatmentGroup','Gender1Female'])['MeanAchievedIndividual'].agg([np.mean, np.std]).reset_index()

# Define a function for a bar plot
def barplot(x_data, y_data, error_data, x_label, y_label, title):
    _, ax = plt.subplots()
# Draw bars, position them in the center of the tick mark on the x-axis
    ax.bar(x_data, y_data, color = '#537caf', align = 'center')
    
# Draw error bars to show standard deviation, set ls to 'none'
# to remove line between points
    ax.errorbar(x_data, y_data, yerr = error_data, color = '#297050', ls = 'none', lw = 2, capthick = 2)
    ax.set_ylabel(y_label)
    ax.set_xlabel(x_label)
    ax.set_title(title)

# Call the function to create a bar plot
barplot(x_data = mean_total_co_day[mean_total_co_day['Gender1Female']=='Yes'].TreatmentGroup.values
        , y_data = mean_total_co_day[mean_total_co_day['Gender1Female']=='Yes']['mean']
        , error_data = mean_total_co_day[mean_total_co_day['Gender1Female']=='Yes']['std']
        , x_label = 'Treatment Group'
        , y_label = 'Mean Achievement'
        , title = 'Mean Achievement for Females with Standard Deviation bars')

plt.xticks([2, 3, 4, 5], ["No SHG, No Prize", "SHG, No Prize", "No SHG, Prize",'SHG, Prize'])
plt.show()

# This figure shows the standard deviation for females in different treatment groups
# We can do the same with males by changing 'Yes' to 'No' in the code above
```

# Report on 5 interesting things about my data 

1) Balance check for age and risk level

By creating boxplots, I looked for outliers in suspected variables. It's important that the students in the different treatments are in the same age group for us to be able to compare them. Repeating a grade was very common in the schools we worked in and it's therefore not certain that they're all in the official 9th grade age group. 
We can see that we only have one real outlier (20 year old), but the rest is balanced.
We can see the same over a histogram plot, showing us the mean and distribution of age over the control group and treatment 4. Comparing different treatments gives a similar histogram.
The reported risk level tells us interestingly, that some students clame they are complete risk averse (0/10) or risk loving (10/10).  

2) Correlation

We can see how the different variables are correlated in a clear way by using heatmaps.
The first heatmap looks at the correlation of variables in our summary statistics. We find that the treatment groups are correlated with higher goal achievment, which is important and the map makes sense in that number of people in the household is highly correlated with having a sibling in school. There are no significant correlation for the other variables.
In the second heatmap, we can see the correlation between the different treatment groups. We notice that the two incentive treatments are correlated. Both of these treatments are correlated with higher mean individual goal achivement, which makes sense and shows how strong the effect of the incentive was.

3) Mean Achievement by treatment groups and gender

Previous analysis in Stata had a graph of mean individual goal achievement by treatment. I included in tables and graphs gender to go more in detail to look at the effect of the different treatments. This enables us to look at mean, variance and standard deviation as well. 
We can see that the highest overall performance was in Treatment 5 (Self-help group and prize), boys and girls are doing equally good in this treatment. There is not a significant difference between Treatment 4 (No self-help group and prize) and Treatment 5, which illustrates the importance of incentives, although the coefficient is higher for Treatment 5. There is however a clear difference between boys and girls, where girls are doing significantly better in the incentive treatment. Interestingly, there is a significant difference between Treatment 2 (No self-help group and no prize) and Treatment 3 (Self-help groups and no prize). Self-help groups alone significantly increase the probability of achieving a goal. Similarly to Treatment 4, girls are performing better than boys in these two treatments. The control group is not represented because they are not selecting goals. 
The swarmplot gives a more detailed decription of this and we can see some clear differences between males and females.

4) Higher Self-esteem may lead to higher goal achievement

The regplot and jointplot shows how there seem to be a positive correlation between self-esteem and individual mean goal achievement. This is interesting in that our treatments may be more effective on students with higher level of self-esteem. This might be important for the treatment effects and is potentially a nice addition to our paper and something needed to be explored further in future regressions.

5) Difficult of goals may play a role in goal achievement when combined with risk and self-esteem

In the multiple tables at the end we can look at how the difficulty level a student give for their objective 
might be correlated with their self-esteem, risk level and the mean achievement of the student.
We can see in the upper right corner, that picking goals with difficulty level 'Hard' seem to be correlated
with stating a high risk level and achiving more goals. This is consistent with goal-setting theory.
We see the opposite for self-esteem, where high self-esteem picks easier goals and achieve them more often.
The histogram shows that most students found their goals to be easy and the goals are achieved regardless of their reported difficulty level.
The residplot shows us that there is no pattern of the residuals of these variables. This is promising for future analysis.
